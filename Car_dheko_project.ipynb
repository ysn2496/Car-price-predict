```python
import pandas as pd
# Load the dataset
file_path = '/content/ML_dataset.csv'
df = pd.read_csv(file_path, low_memory=False)
```

```python
df
```

```python
# Assuming df is your DataFrame and 'price' is your target variable
X = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['bt'] = le.fit_transform(X['bt'])
X['Fuel Type'] = le.fit_transform(X['Fuel Type'])
X['Transmission'] = le.fit_transform(X['Transmission'])
X['Color'] = le.fit_transform(X['Color'])
X['city'] = le.fit_transform(X['city'])
X['Steering Type'] = le.fit_transform(X['Steering Type'])
X['Insurance Validity'] = le.fit_transform(X['Insurance Validity'])
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# prompt: write the code for linear regression using df dataset

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a Linear Regression model
lr_model = LinearRegression()

# Train the model
lr_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = lr_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# prompt: write the code for decision trees

from sklearn.tree import DecisionTreeRegressor

# Create a Decision Tree Regression model
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model
dt_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = dt_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# prompt: write the code for random forest

from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regression model
rf_model = RandomForestRegressor(random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# prompt: write the code for gradient boosting

from sklearn.ensemble import GradientBoostingRegressor

# Create a Gradient Boosting Regression model
gb_model = GradientBoostingRegressor(random_state=42)

# Train the model
gb_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = gb_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# prompt: write the code for hyperparameter tuning of the rf model

from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a Random Forest Regression model
rf_model = RandomForestRegressor(random_state=42)

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print(f'Best parameters: {best_params}')

# Get the best score
best_score = grid_search.best_score_
print(f'Best score: {best_score}')

# Train the model with the best parameters
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = best_rf_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# prompt: Regularization: Apply regularization techniques to prevent overfitting.
# Lasso (L1) and Ridge (L2) regularization.

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

# Lasso Regression
lasso_model = Lasso(alpha=0.1)  # Adjust alpha for regularization strength
lasso_model.fit(X_train, y_train)
y_pred_lasso = lasso_model.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
print(f'Lasso - Mean Squared Error: {mse_lasso}')
print(f'Lasso - R-squared: {r2_lasso}')

# Ridge Regression
ridge_model = Ridge(alpha=0.1)  # Adjust alpha for regularization strength
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f'Ridge - Mean Squared Error: {mse_ridge}')
print(f'Ridge - R-squared: {r2_ridge}')
```

```python
# prompt: how to find the model is overfitting or underfitting by using ridge and lasso method

# Calculate the difference between training and testing R-squared values
train_r2_lasso = lasso_model.score(X_train, y_train)
test_r2_lasso = lasso_model.score(X_test, y_test)
diff_lasso = train_r2_lasso - test_r2_lasso

train_r2_ridge = ridge_model.score(X_train, y_train)
test_r2_ridge = ridge_model.score(X_test, y_test)
diff_ridge = train_r2_ridge - test_r2_ridge

print(f"Lasso - Difference between train and test R-squared: {diff_lasso}")
print(f"Ridge - Difference between train and test R-squared: {diff_ridge}")

# Analyze the differences
if diff_lasso > 0.1:  # Adjust the threshold as needed
    print("Lasso model might be overfitting.")
elif diff_lasso < -0.1:
    print("Lasso model might be underfitting.")
else:
    print("Lasso model seems to have a good fit.")

if diff_ridge > 0.1:  # Adjust the threshold as needed
    print("Ridge model might be overfitting.")
elif diff_ridge < -0.1:
    print("Ridge model might be underfitting.")
else:
    print("Ridge model seems to have a good fit.")
```
