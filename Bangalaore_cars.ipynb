```python
import pandas as pd
import json
import re

# Load the Excel file and the specific sheet
file_path = 'D:/Project 3 Cardheko/bangalore_cars.xlsx'
df = pd.read_excel(file_path, sheet_name='bangalore_cars.csv')

# Function to convert Python-style dictionary strings to valid JSON
def convert_to_json_format(text):
    text = text.replace("'", '"').replace("None", "null").replace("True", "true").replace("False", "false")
    text = re.sub(r'(?<!")(\b\w+\b)(?=:)', r'"\1"', text)
    return text

# Apply the conversion to the 'new_car_detail' column
df['new_car_detail'] = df['new_car_detail'].apply(convert_to_json_format)

# Function to flatten the JSON data
def flatten_json_column(df, column_name):
    return pd.json_normalize(df[column_name].apply(json.loads))

# Flatten the 'new_car_detail' column
flattened_details = flatten_json_column(df, 'new_car_detail')

# Display the structured data
flattened_details.head()


```

```python
#Optionally, save the structured data to a new CSV file
flattened_details.to_csv('structured_new_car_details.csv', index=False)
```

```python
import pandas as pd
import json

# Load the Excel file
file_path = 'D:/Project 3 Cardheko/bangalore_cars.xlsx'
df = pd.read_excel(file_path)

# Clean and parse JSON-like strings in 'new_car_feature'
df['new_car_feature'] = df['new_car_feature'].apply(lambda x: json.loads(x.replace("'", '"').replace('None', '"1"')))

# Function to flatten the 'new_car_feature' column
def flatten_car_feature(feature):
    # Extract 'top' features
    top_features = [item['value'] for item in feature.get('top', [])]
    
    # Extract 'data' features with headings and subheadings
    data_features = []
    for group in feature.get('data', []):
        heading = group['heading']
        sub_heading = group['subHeading']
        for item in group.get('list', []):
            data_features.append({
                'heading': heading,
                'sub_heading': sub_heading,
                'feature': item['value']
            })
    
    return {'top_features': top_features, 'data_features': data_features}

# Apply the function to extract and flatten data from 'new_car_feature'
df['new_car_feature_flat'] = df['new_car_feature'].apply(flatten_car_feature)

# Separate the 'top_features' and 'data_features' into different DataFrames
top_features_list = []
data_features_list = []

for index, row in df.iterrows():
    top_features = row['new_car_feature_flat']['top_features']
    data_features = row['new_car_feature_flat']['data_features']
    
    # Add the index to keep track of rows
    top_features_list.extend([{'row_index': index, 'feature_type': 'top', 'Feature': feature} for feature in top_features])
    data_features_list.extend([{'row_index': index, 'feature_type': 'data', 'heading': feature['heading'], 'sub_heading': feature['sub_heading'], 'feature': feature['feature']} for feature in data_features])

# Create DataFrames for 'top_features' and 'data_features'
top_features_df = pd.DataFrame(top_features_list)
data_features_df = pd.DataFrame(data_features_list)

# Combine 'top_features' and 'data_features' into a single DataFrame
combined_df = pd.concat([top_features_df, data_features_df], axis=0)

# Create unique identifiers for pivoting
combined_df['feature_id'] = combined_df.groupby(['row_index', 'feature_type']).cumcount() + 1

# Pivot the DataFrame
pivot_df = combined_df.pivot_table(
    index=['row_index', 'feature_type'],
    columns='feature_id',
    values='Feature' if 'Feature' in combined_df.columns else 'feature',
    aggfunc='first'  # Use 'first' in case of duplicates, adjust if needed
).reset_index()

# Display the column names before renaming
print("Column names before renaming:")
print(pivot_df.columns)

# Rename columns for better readability
new_column_names = []
for col in pivot_df.columns:
    if isinstance(col, tuple):
        new_column_names.append(f'{col[0]}_{col[1]}')
    else:
        new_column_names.append(col)

pivot_df.columns = new_column_names

# Display the resulting DataFrame
print("Pivoted DataFrame:")
print(pivot_df.head())


```

```python
import pandas as pd
import json

# Load the Excel file
file_path = 'D:/Project 3 Cardheko/bangalore_cars.xlsx'
df = pd.read_excel(file_path)

# Clean and parse JSON-like strings in 'new_car_feature'
df['new_car_feature'] = df['new_car_feature'].apply(lambda x: json.loads(x.replace("'", '"').replace('None', '1')))

# Function to flatten the 'new_car_feature' column
def flatten_car_feature(feature):
    # Extract 'commonIcon'
    common_icon = feature.get('commonIcon', None)
    
    # Extract 'top' features
    top_features = {f"Top Feature {i+1}": item['value'] for i, item in enumerate(feature.get('top', []))}
    
    # Extract 'data' features with headings and subheadings
    data_features = {}
    for group in feature.get('data', []):
        heading = group['heading']
        sub_heading = group['subHeading']
        for i, item in enumerate(group.get('list', [])):
            key = f"{heading}_{sub_heading}_Feature {i+1}"
            data_features[key] = item['value']
    
    return {**top_features, **data_features, 'commonIcon': common_icon}

# Apply the function to extract and flatten data from 'new_car_feature'
df['flattened_features'] = df['new_car_feature'].apply(flatten_car_feature)

# Convert the flattened features into a DataFrame
features_df = pd.DataFrame(df['flattened_features'].tolist())

# Concatenate the original DataFrame with the new features DataFrame
result_df = pd.concat([df.drop(columns=['new_car_feature', 'flattened_features']), features_df], axis=1)

# Display the resulting DataFrame
print("Final DataFrame with all features in columns, including commonIcon:")
print(result_df.head())

```

```python
import pandas as pd
import json

# Load the Excel file
file_path = 'D:/Project 3 Cardheko/bangalore_cars.xlsx'
df = pd.read_excel(file_path)

# Clean and parse JSON-like strings in 'new_car_specs'
df['new_car_specs'] = df['new_car_specs'].apply(lambda x: json.loads(x.replace("'", '"').replace('None', '1')))

# Function to flatten the 'new_car_specs' column
def flatten_car_specs(specs):
    # Initialize a dictionary to store the flattened data
    flattened_data = {}
    
    # Extract 'commonIcon'
    flattened_data['commonIcon'] = specs.get('commonIcon', None)
    
    # Extract 'top' specifications
    for item in specs.get('top', []):
        key = item.get('key')
        value = item.get('value')
        if key:  # Ensure key is not None
            flattened_data[key] = value
    
    # Extract 'data' features with headings and subheadings
    for group in specs.get('data', []):
        for item in group.get('list', []):
            key = item.get('key')
            value = item.get('value')
            if key:  # Ensure key is not None
                flattened_data[key] = value
    
    return flattened_data

# Apply the function to extract and flatten data from 'new_car_specs'
df['flattened_specs'] = df['new_car_specs'].apply(flatten_car_specs)

# Convert the flattened specs into a DataFrame
specs_df = pd.DataFrame(df['flattened_specs'].tolist())

# Concatenate the original DataFrame with the new specs DataFrame
result_df = pd.concat([df.drop(columns=['new_car_specs', 'flattened_specs']), specs_df], axis=1)

# Display the resulting DataFrame
print("Final DataFrame with keys as column names and values as rows:")
print(result_df.head())


```

```python
import pandas as pd
import re

# Load the dataset
file_path = 'D:/Project 3 Cardheko/structured_new_car_details.csv'  # Update this path if running locally
df = pd.read_csv(file_path)

# Drop columns with all missing data or single unique values
df_cleaned = df.drop(columns=['priceSaving', 'priceFixedText', 'trendingText.imgUrl', 'trendingText.heading', 'trendingText.desc'])

# Function to clean and convert 'price' values
def clean_price(price):
    if pd.isna(price):
        return None
    # Remove currency symbols and words like 'Lakh', 'Crore'
    price = price.replace('₹', '').replace('Lakh', '').replace('Crore', '')
    # Remove any remaining non-numeric characters
    price = re.sub(r'[^\d.]', '', price)
    # Convert to float
    return float(price) * 1e5 if 'Lakh' in price else (float(price) * 1e7 if 'Crore' in price else float(price))

# Clean 'price' column
df_cleaned['price'] = df_cleaned['price'].apply(clean_price)

# Clean and convert 'km' column (remove commas and convert to float)
df_cleaned['km'] = df_cleaned['km'].str.replace(',', '').astype(float)

# Check for missing values
missing_data = df_cleaned.isnull().sum()

# Display the cleaned DataFrame and missing data info
print(df_cleaned.head())
print(missing_data)


```
