```python
import pandas as pd
# Load the dataset
file_path = 'D:/Project 3 Cardheko/Car_dheko.csv'
df = pd.read_csv(file_path, low_memory=False)
```

```python
import seaborn as sns
```

```python
sns.distplot(df['Mileage'])
```

```python
sns.boxplot(df['Mileage'])
```

```python
#Finding the IQR-Inter Quartile Range
percentile25 = df['Mileage'].quantile(0.25)
percentile75 = df['Mileage'].quantile(0.75)
iqr=percentile75 - percentile25
print(iqr)
print(percentile25)
print(percentile75)
```

```python
#Finding upper and lower limit
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print(upper_limit)
print(lower_limit)
```

```python
#finding outliers
df[df['Mileage'] > upper_limit]
```

```python
#finding outliers
df[df['Mileage'] < lower_limit]
```
